```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import json
import math
import pandas as pd
import openpyxl
```

```python
data = {'Title': [],
        'Writer': [],
        'Price': [],
       'URL':[],
       'ISBN': [],
       'Publisher': []
       }

inp = input()
# inp = "https://www.ianos.gr/books/epistimes/astronomia"

# if "ianos.gr" in inp:

driver = webdriver.Chrome()
driver.get(inp)
elements = driver.find_elements(By.CLASS_NAME, "toolbar-number")
results_per_page = float(elements[1].text)
total_results = float(elements[2].text)
number_of_pages = math.ceil(total_results / results_per_page)
print(f"The number of pages is {number_of_pages}")

my_list = []
# for i in range (1,2):
for i in range (1,number_of_pages+1):
    next_page_url = inp + f"?p={i}"
    driver.get(next_page_url)
    print(i)
    elements = driver.find_elements(By.CLASS_NAME, "product-item")
    elements = elements[:int(results_per_page)]

    try:
        for element in elements:
            writer = element.get_attribute("data-gtm-brand")
            price = float(element.get_attribute("data-price")) - float(element.get_attribute("data-discount")) if element.get_attribute("data-discount") else float(element.get_attribute("data-price"))
            title = element.find_element(By.XPATH, ".//div//div//strong//a").text
            url = element.find_element(By.XPATH, ".//div//div//strong//a").get_attribute("href")
            data['Writer'].append(writer)
            data['Price'].append(str(round(price,2)))
            data['Title'].append(title)
            data['URL'].append(url)
    except:
        print("Error")
    finally:
        print("Found all books")

books = data['URL']
for book in books: 
    driver.get(book)
    print(book)
    table = driver.find_element(By.TAG_NAME, "table")
    try:
        isbn = table.find_element(By.XPATH, ".//tr//td[@data-th='ISBN']").text
    except:
        isbn = ""
    data['ISBN'].append(isbn)
    try:
        publisher = table.find_element(By.XPATH, ".//tr//td[@data-th='Εκδότης']").text
    except:
        publisher = ""
    data['Publisher'].append(publisher)

print("Data collection is done")

driver.quit()
df = pd.DataFrame(data)
# Save the DataFrame to an Excel file]
df.to_excel('outputIanos.xlsx', index=False)
print("File saved")
```
